## 3. Analyse Data - MVP
For the MVP I'll analyse the sub-groups of people with a mental health condition.
As the criteria for having a MH condition I'll consider the "treatment" entry, as there isn't any other one especified for this.

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('survey_clean.csv', index_col=0)
data.head()
```

```python
data['treatment'].value_counts()
```

```python
data['treatment'].isnull().sum()
```

#### 3.1 MH condition by Country
I'll manipulate the data to get a table with the percentage of 'Yes' and 'No' for the MH Condition (using the treatment), by country and weighted with the number of records per country.

```python
country_total = data[['Country','treatment']].groupby('Country').count()
country_total.reset_index(inplace=True)
country_total.columns = ['Country', 'Total']
country_total
```

```python
country_yes = data[['Country','treatment']][data['treatment'] == 'Yes'].groupby('Country').count()
country_yes.reset_index(inplace=True)
country_yes.columns = ['Country', 'Condition Yes']
country_yes
```

```python
country_no = data[['Country','treatment']][data['treatment'] == 'No'].groupby('Country').count()
country_no.reset_index(inplace=True)
country_no.columns = ['Country', 'Condition No']
country_no
```

```python
condition_by_country = pd.merge(country_yes, country_no, on='Country')
condition_by_country = pd.merge(condition_by_country, country_total, on='Country')
condition_by_country
```

```python
condition_by_country['% Yes'] = round(condition_by_country['Condition Yes'] / condition_by_country['Total'] * 100, 1)
condition_by_country['% No'] = round(condition_by_country['Condition No'] / condition_by_country['Total'] * 100, 1)
condition_by_country.sort_values(by=['% Yes'], ascending=False, inplace=True)
condition_by_country
```

```python
cy = plt.bar(condition_by_country['Country'], condition_by_country['% Yes'], color='c')
cn = plt.bar(condition_by_country['Country'], condition_by_country['% No'], bottom=condition_by_country['% Yes'], color='lavender')
plt.xticks(rotation=45)
plt.ylabel('% Records')
plt.title('MH Condition by Country')
plt.legend((cy,cn),('Yes','No'))
plt.show()
```

Conclusion: from this chart we see the countries with more MH condition (Australia) and less MH condition (Netherlands), we also see that in general about 50% of the people have a MH condition which is a very significant insight.

#### 3.2 MH condition by Age
I'll manipulate the data to get a table with the percentage of 'Yes' and 'No' for the MH Condition (using the treatment), by age and weighted with the number of records per age.

```python
age_total = data[['Age','treatment']].groupby('Age').count()
age_total.reset_index(inplace=True)
age_total.columns = ['Age', 'Total']
age_total = age_total.astype(int)
```

```python
age_yes = data[['Age','treatment']][data['treatment'] == 'Yes'].groupby('Age').count()
age_yes.reset_index(inplace=True)
age_yes.columns = ['Age', 'Condition Yes']
age_yes = age_yes.astype(int)
```

```python
age_no = data[['Age','treatment']][data['treatment'] == 'No'].groupby('Age').count()
age_no.reset_index(inplace=True)
age_no.columns = ['Age', 'Condition No']
age_no = age_no.astype(int)
```

```python
condition_by_age = pd.merge(age_yes, age_no, on='Age')
condition_by_age = pd.merge(condition_by_age, age_total, on='Age')
```

```python
condition_by_age['% Yes'] = round(condition_by_age['Condition Yes'] / condition_by_age['Total'] * 100, 1)
condition_by_age['% No'] = round(condition_by_age['Condition No'] / condition_by_age['Total'] * 100, 1)
condition_by_age
```

```python
ay = plt.bar(condition_by_age['Age'], condition_by_age['% Yes'], color='c')
an = plt.bar(condition_by_age['Age'], condition_by_age['% No'], bottom=condition_by_age['% Yes'], color='lavender')
plt.ylabel('% Records')
plt.xlabel('Age (years)')
plt.title('MH Condition by Age')
plt.legend((ay,an),('Yes','No'))
plt.show()
```

```python
# the previous chart doesn't show clear conclusions so I'll groups the ages by ranges using bins
condition_by_age['Age bins'] = pd.cut(condition_by_age['Age'], 5, labels=['18-26','27-34','35-43','44-49','56-60'])
condition_by_age
```

```python
condition_by_age_bins = condition_by_age.groupby('Age bins').mean()
condition_by_age_bins.reset_index(inplace=True)
condition_by_age_bins
```

```python
ayb = plt.bar(condition_by_age_bins['Age bins'], condition_by_age_bins['% Yes'], color='c')
anb = plt.bar(condition_by_age_bins['Age bins'], condition_by_age_bins['% No'], bottom=condition_by_age_bins['% Yes'], color='lavender')
plt.ylabel('% Records')
plt.xlabel('Age (years)')
plt.title('MH Condition by Age')
plt.legend((ayb,anb),('Yes','No'))
plt.show()
```

Conclusion: from this chart we see an slightly increasing tendency of MH condition with age.
